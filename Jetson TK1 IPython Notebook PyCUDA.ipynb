# NVIDIA Jetson TK1でIPython NotebookとPyCUDAを使ってみる

Jetson TK1の上でIPython notebookとPyCUDAを使うことができたので、IPythonとPyCUDAの紹介も兼ねて、方法をまとめておきます。

このメモはIPython notebook形式です。右上のボタンからダウンロードすることで、実行サンプルを実際にJetson TK1のIPython notebookで動作させてみることが可能です。是非お試しください。

## IPython notebookをJetson TK1にインストールする

ネットワークに接続されている状態で、コマンドラインからapt-getでインストールします。依存関係は自動的に解決されますので、下記で必要なモジュールがインストールされます。

    $ sudo apt-get install ipython-notebook python-matplotlib

インストールが成功したらIPython Notebookを起動してみます。他の端末から利用したいので、ブラウザを自動で開かないように、またローカルからしかアクセスできないようデフォルトではIPが127.0.0.1になっていますが、外部からのアクセスができるよう0.0.0.0(INADDR_ANY)を明示的に指定します。

    $ ipython notebook --pylab inline --no-browser --ip=0.0.0.0

ブラウザで http://{IP-address}:8888/ を開きます。IPython Notebook Dashboardが出てくるはずです。New Notebookボタンを押して、新しいノートを作って試してみます。適当な関数をプロットさせてみて、グラフがインラインで描画できたら、numpyならびに matplotlibは正常に動作しています。

```python
x = arange(0, 10, 0.1)
plot(sin(x))
```

## PyCUDAをJetson TK1にインストールする

PyCUDAのインストールに先立って、まずCUDA SDKがインストールされている必要があります。バージョンは6.0で試しました。CUDA SDKのインストールについてはこちらを参考にしてください。https://developer.nvidia.com/jetson-tk1-support

PyCUDAのインストールはUbuntuでの公式な手順 http://wiki.tiker.net/PyCuda/Installation/Linux/Ubuntu に従いますが、ファイル名やパス名に若干修正が必要でした。

まずビルドに必要なライブラリをインストールします。

    $ sudo apt-get install build-essential python-dev python-setuptools libboost-python-dev libboost-thread-dev -y

続いて、PyCUDAのパッケージをダウンロードし展開します。

    $ wget https://pypi.python.org/packages/source/p/pycuda/pycuda-2014.1.tar.gz
$ tar xfz pycuda-2014.1.tar.gz
    
configureします。注意点としては、cuda-rootに /usr/local/cuda を指定することと、libcuda.soが /usr/lib/arm-linux-gnueabihf/tegra にあるので、それを明示的に指定が必要なことです。またboostのライブラリが-mtだとエラーになりますので、-mt無しを指定します(もしかしたらboost関係のオプション指定は不要かもしれません)。

    $ cd pycuda-2014.1
$ ./configure.py --cuda-root=/usr/local/cuda --cudadrv-lib-dir=/usr/lib/arm-linux-gnueabihf/tegra --boost-inc-dir=/usr/include --boost-lib-dir=/usr/lib --boost-python-libname=boost_python --boost-thread-libname=boost_thread --no-use-shipped-boost
    
make、そしてsetup.py installします。

    $ make -j 4
$ sudo python setup.py install

実行時にpytoolsが必要なので追加でインストールしておきます。

    $ sudo apt-get install python-pytools

インストールに成功したら、いよいよIPython NotebookでPyCUDAを試してみましょう。

## IPython NotebookでPyCUDAを試す

先ほどと同じようにIPython Notebookを起動し、ブラウザから開いておきます。そして新しくノートブックを作ります。

     $ ipython notebook --pylab inline --no-browser --ip=0.0.0.0

Cellに下記を入力し、Ctrl-Enterで実行してみます。

```python
import pycuda
import pycuda.autoinit
import pycuda.driver as drv
from pycuda.compiler import SourceModule
```

これがうまくいったら、まずは第一段階突破、IPythonへのPyCUDAのロードは成功です。

続いてCUDAのコードです。コードは文字列としてSourceModuleに渡します。小規模な例として、みんな大好きジュリア集合をやってみます。私にとっても始めてのCUDAなので、コードはこちらのサイトを参考にしました。http://d.hatena.ne.jp/norio515/20111112/1321105385

```python
mod = SourceModule("""
__device__ char julia(int x, int y, int w, int h)
{
    const float scale = 1;
    float jx = scale * (float)(w / 2 - x) / (w / 2);
    float jy = scale * (float)(h / 2 - y) / (h / 2);
    const float cx = -0.8;
    const float cy = 0.156;
    int i = 0;
    for (i = 0; i < 255; i++) {
        float ax = jx * jx - jy * jy + cx;
        float ay = 2 * jx * jy + cy;
        if (ax * ax + ay * ay > 1000) {
            return i;
        }
        jx = ax;
        jy = ay;
    }
    return 0;
}

__global__ void kernel(char *ptr)
{
    int x = blockIdx.x;
    int y = blockIdx.y;
    int w = gridDim.x;
    int h = gridDim.y;
    int i = x + y * w;
    ptr[i] = julia(x, y, w, h);
}
""")
```

Ctrl-EnterでRunすると、コードをnvccでコンパイルしますので、数秒の時間を要します。ここがうまく行けば第二段階突破です。

続いて上記のコードを呼び出すpythonの関数を作ります。やっていることは、numpyの配列と同じサイズの領域をGPU側に確保し、先ほどのコードをPythonの関数として呼び出し、GPUでの計算が終わったら結果の配列をホスト側にコピーしていることがわかります。

```python
def julia(size):
    w, h = size
    data = np.zeros((h,w), np.uint8)
    mem = drv.mem_alloc(data.nbytes)
    kernel = mod.get_function("kernel")
    kernel(mem, block=(1, 1, 1), grid=size)
    drv.memcpy_dtoh(data, mem)
    return data
```

いよいよ実行してみます。引数sizeには、*width*, *height*を列で与えます。ついでに%timeキーワードを付けて実行時間を計測してみます。

```python
%time d = julia((512, 512))
```

変数dに入った結果は、指定されたサイズを持つuint8の配列です。imshow()でそのまま表示することができます。

```python
imshow(d)
```

## 実行速度をCPUによる計算と比べてみる

CPUで計算した場合と比較してみます。同じくジュリア図形のPythonによる実装です。

```python
def julia_cpu(size):
    w, h = size
    data = np.zeros((h,w), np.uint8)
    def julia(x, y):
        scale = 1
        jx = scale * (w / 2.0 - x) / (w / 2.0)
        jy = scale * (h / 2.0 - y) / (h / 2.0)
        c = complex(-0.8, 0.156)
        a = complex(jx, jy)
        for i in xrange(255):
            a = a * a + c
            if a.real*a.real + a.imag*a.imag > 1000:
                return i
        return 0
    for y in xrange(h):
        for x in xrange(w):
            data[y][x] = julia(x, y)
    return data
```

```python
%time c = julia_cpu((512,512))
imshow(c)
```

同じ図が得られました。NVIDIA Tegra K1のCortex-A15での結果ということになりますが、10秒程度かかっています。

なぜかCUDA版は時折爆速なことがあり、計測のぶれが大きいのですが、CUDA版はCPU版にくらべておよそ一桁速いようです。特に2のべきのサイズだと爆速なことが多いのですが必ずではありません。本来ならCPU版はマルチスレッドで実装して比較すべきかもしれませんが、ここは簡便に済ませました。

## マンデルブロ集合も試してみる

ついでに、こちらもみんな大好きマンデルブロ集合もPyCUDAで計算してみます。

CUDAのコードを書いて、

```python
mod_mb = SourceModule("""
__device__ char mandelbrot(int x, int y, int w, int h)
{
    const float scale = 1;
    float cx = scale * (float)(x - 3 * w / 4) / (w / 2);
    float cy = scale * (float)(y - h / 2) / (h / 2);
    float jx = 0;
    float jy = 0;
    int i;
    for (i = 0; i < 256; i++) {
        float ax = jx * jx - jy * jy + cx;
        float ay = 2 * jx * jy + cy;
        if (ax * ax + ay * ay > 4.0) {
            return i;
        }
        jx = ax;
        jy = ay;
    }
    return 0;
}

__global__ void kernel_mb(char *ptr)
{
    int x = blockIdx.x;
    int y = blockIdx.y;
    int w = gridDim.x;
    int h = gridDim.y;
    int i = x + y * w;
    ptr[i] = mandelbrot(x, y, w, h);
}
""")
```

Pythonの関数を作って、

```python
def mandelbrot(size):
    w, h = size
    data = np.zeros((h,w), np.uint8)
    mem = drv.mem_alloc(data.nbytes)
    kernel = mod_mb.get_function("kernel_mb")
    kernel(mem, block=(1, 1, 1), grid=size)
    drv.memcpy_dtoh(data, mem)
    return data
```

そして実行します。

```python
%time m = mandelbrot((512, 512))
imshow(m)
```

ちゃんと動きました。

IPython notebookの環境は、試行錯誤するのに適した環境だと思います。コードやパラメータを少しづつ変えて、結果を並べて比べたりなどが簡単にできます。他の目的で開発する場合でも、データやコードの断片をIPython notebookで十分検討し練り上げてから、ターゲットの環境にコードを移していくようなプロセスを踏むことで、効率的かつ創発的に開発を進めることができるのではないでしょうか。しかもその結果をgithubに上げておくだけでnbviewerで簡単に公開するもできてしまいます。是非おススメしたいです。
